## Exploratory Data Analysis

```python
#Get basic information about dataset
df.info()
```

```python
#Drop columns that doesn't provide useful information
df.drop(['Unnamed: 0', 'index'], axis = 1, inplace = True)
```

```python
df['Revenue'].value_counts()
```

```python
# Change -1 to False in our Easy Apply feature
df['Easy Apply'] = np.where(df['Easy Apply'] == '-1', 'False', df['Easy Apply'])
df['Easy Apply']
```

```python
# Format the Company Name Feature into readable format
df['Company Name'] = df['Company Name'].apply(lambda x: x[:-4] if '\n' in x else x)
df['Company Name']
```

```python
df['Size'].value_counts()
```

```python
#Get the number of each job titles
df['Job Title'].value_counts()
```

```python
df['Salary Estimate'].value_counts()
```

```python
#Create a new column to calculate the average salary of each job 
def average_salary(x):
  col = x['Salary Estimate']
  if 'Per Hour' in col:
    lower_limit = int(col.split(' ')[0].split('-')[0][1:]) * 40 * 52
    upper_limit = int(col.split(' ')[0].split('-')[1][1:]) * 40 * 52
    return (lower_limit + upper_limit) / 2
  else:
    lower_limit = int(col.split(' ')[0].split('-')[0][1:-1]) * 1000
    upper_limit = col.split(' ')[0].split('-')[1]
    if '(' in upper_limit:
      ind = upper_limit.index('(')
      upper_limit = int(upper_limit[1:ind-1]) * 1000
    else:
      upper_limit = int(upper_limit[1:-1]) * 1000
    return (lower_limit + upper_limit) / 2

df['average_salary'] = df.apply(average_salary, axis = 1)
```

```python
#Change rating with value of -1 to 0
df['Rating'] = np.where(df['Rating'] == -1.0, 0, df['Rating'])
```

```python
# Most popular jobs in the dataset
jobs = df['Job Title'].value_counts()
most_popular_jobs = jobs[jobs >= 40]
most_popular_jobs 
```

```python
#Most popular industries
industry = df['Industry'].value_counts()
popular_industries = industry[industry >= 100]
popular_industries.drop('-1', axis=0, inplace = True)
popular_industries
```

```python
#Average salary of the most popular jobs in the dataset
list_high_pay_jobs = list(most_popular_jobs.index)
pop_jobs = df[(df['Job Title'] == 'Data Scientist')| (df['Job Title'] == 'Data Engineer') |(df['Job Title'] == 'Data Analyst')|
             (df['Job Title'] == 'Senior Data Scientist')| (df['Job Title'] == 'Senior Data Analyst')|(df['Job Title'] == 'Machine Learning Engineer')|
             (df['Job Title'] == 'Big Data Engineer')]

highest_paying_jobs = pop_jobs.groupby('Job Title').mean()['average_salary']
highest_paying_jobs
```

```python
#Create a categorical feature from Rating feature
df['Rating_cat'] = pd.cut(df['Rating'], bins = 5, labels = ['0-1', '1-2', '2-3', '3-4', '4-5'])
df['Rating_cat'].value_counts()
```

## Visualisations

```python
#Visualisation to show most popular Data Science Jobs in our dataset
def pop_data_jobs():
  plt.figure(figsize=(7,5))
  plt.title('Most Popular Data Science Jobs', size = 14)
  sns.set_style('whitegrid')
  count_plot = sns.countplot(y = 'Job Title', data = pop_jobs, palette = 'Purples')
  plt.close()
  return count_plot.figure
pop_data_jobs()
```

```python
 # Scatter plot
 def scatter():
  plt.figure(figsize = (7,5))
  plt.title('Ratings vs Average salary')
  plot = sns.scatterplot(x = 'Rating', y = 'average_salary', data = pop_jobs, hue = 'Job Title')
  plt.legend(loc='upper left')
  plt.close()
  return plot.figure

scatter()
```

```python
#First five locations for jobs
loc_jobs = pop_jobs['Location'].value_counts().head()
loc_val = loc_jobs.values
loc_ind = loc_jobs.index
total_sum = sum(pop_jobs['Location'].value_counts().values)
percentages = np.round((loc_val / total_sum) * 100, 1)

#Bar Chart
def pie():
  plt.figure()
  plt.figure(figsize=(7,5))
  plt.title('Most Popular locations for Data Science Jobs', size = 14)
  bar_plot = sns.barplot(x = loc_ind, y = loc_val, palette = 'Purples')
  patches = bar_plot.patches
  for i in range(len(patches)):
    x = patches[i].get_x() + patches[i].get_width()/2
    y = patches[i].get_height()+.05
    bar_plot.annotate('{:.1f}%'.format(percentages[i]), (x, y), ha='center')
    plt.close()
  return bar_plot.figure

plot_pie = pie()
plot_pie
```

```python
pop_jobs.groupby(['Job Title', 'Company Name']).mean()
```

```python
#Creates bar chart which highlights company with highest rating based on specific jobs
symbols = list(most_popular_jobs.index)
select = pn.widgets.Select(name = 'Select', options = symbols)
stem = pop_jobs.groupby(['Job Title', 'Company Name']).mean()
@pn.depends(select)
def average_plot(symbol):
  plt.figure(figsize = (7,5))
  plt.title('Companies With The Highest Ratings', size = 14)
  BDE = stem.loc[symbol]['Rating'].nlargest(5)
  fig = sns.barplot(y = BDE.index, x = BDE.values, palette = 'Purples')
  plt.close()
  return fig.figure

pn.Column(select, average_plot)
```

<a name = "dashboard"></a>
#Dashboard

```python
title = '# Data Science Jobs Dashboard'
```

```python
Dashboard = pn.WidgetBox(pn.Column(pn.Row(title, select), pn.Row(average_plot, scatter), pn.Row(pop_data_jobs, pie),
                                   align = 'start', sizing_mode = 'stretch_width'))
Dashboard 
```
